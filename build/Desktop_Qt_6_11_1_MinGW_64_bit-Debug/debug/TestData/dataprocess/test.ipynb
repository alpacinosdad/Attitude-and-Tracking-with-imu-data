```python
import pandas as pd
import matplotlib.pyplot as plt
import numpy as np

# ==============================================================================
# 画 3 轴加速度滤波后 + 只看 75~85 秒
# ==============================================================================
def plot_accel_3axis_filtered_75_to_85(df):
    # 时间轴：20ms 一个点
    time_sec = np.arange(0, len(df)*0.02, 0.02)[:len(df)]
    
    # 创建宽松大图
    plt.figure(figsize=(14, 6))

    # 画三条滤波后曲线
    plt.plot(time_sec, df['Accel_x_h'], color='blue', linewidth=2, label='Accel X (filtered)')
    plt.plot(time_sec, df['Accel_y_h'], color='green', linewidth=2, label='Accel Y (filtered)')
    plt.plot(time_sec, df['Accel_z_h'], color='red', linewidth=2, label='Accel Z (filtered)')

    # ✅ 只显示 75s ~ 85s
    plt.xlim(0, 20)

    plt.title('Accelerometer 3-axis Filtered (75s - 85s)', fontsize=14)
    plt.xlabel('Time (s)', fontsize=12)
    plt.ylabel('Acceleration', fontsize=12)
    plt.grid(True, alpha=0.3)
    plt.legend(fontsize=12)
    plt.tight_layout()
    plt.show()

# ==============================================================================
# 读取数据并绘图
# ==============================================================================
df = pd.read_csv("BAT_Heat_Log_Data_2026_05_27_22_00_56.csv")
plot_accel_3axis_filtered_75_to_85(df)
```

```python
import matplotlib
matplotlib.use('TkAgg') # 强制使用 Tkinter 弹出独立窗口
import matplotlib.pyplot as plt
import pandas as pd
import numpy as np
from matplotlib.animation import FuncAnimation

# ... 下面继续接你之前的代码 ...

# ==========================================
# 1. 读取你的表格数据
# ==========================================
# 请将 'your_data.csv' 替换为你实际的表格文件路径
# 如果是 excel 文件，请使用 pd.read_excel('your_data.xlsx')
file_path = 'BAT_Heat_Log_Data_2026_05_27_21_01_55_imu_result.csv' 
df = pd.read_csv(file_path)

# 为了防止数据过长导致动画卡顿，如果数据量超过几万行，可以采取降采样(每隔N行取一次)
# df = df.iloc[::2].reset_index(drop=True) 

# 获取总帧数
num_frames = len(df)

# 根据 Time_s 计算平均每帧的时间间隔 (毫秒)，用于控制真实播放速度
if num_frames > 1:
    dt_seconds = df['Time_s'].iloc[1] - df['Time_s'].iloc[0]
    interval_ms = max(1, int(dt_seconds * 1000)) # 转换为毫秒
else:
    interval_ms = 50

# ==========================================
# 2. 设置 3D 可视化场景
# ==========================================
fig = plt.figure(figsize=(10, 8))
ax = fig.add_subplot(111, projection='3d')

# 设置坐标轴范围 (这是一个归一化长度的参考系)
ax.set_xlim([-1.2, 1.2])
ax.set_ylim([-1.2, 1.2])
ax.set_zlim([-1.2, 1.2])

# 设置坐标轴标签
ax.set_xlabel('Global X')
ax.set_ylabel('Global Y')
ax.set_zlabel('Global Z')

# 初始化三根线，分别代表设备的 X(前)、Y(右)、Z(下) 轴
# 红色代表机头前方，绿色代表右侧，蓝色代表下方（或上方，视你的传感器定义而定）
x_axis, = ax.plot([], [], [], color='red', linewidth=4, label='Device X (Forward)')
y_axis, = ax.plot([], [], [], color='green', linewidth=4, label='Device Y (Right)')
z_axis, = ax.plot([], [], [], color='blue', linewidth=4, label='Device Z (Up/Down)')

# 在左上角显示当前时间
time_text = ax.text2D(0.05, 0.95, '', transform=ax.transAxes, fontsize=12)

ax.legend()

# 基础坐标系的三个单位向量
base_x = np.array([1.0, 0.0, 0.0])
base_y = np.array([0.0, 1.0, 0.0])
base_z = np.array([0.0, 0.0, 1.0])

# ==========================================
# 3. 动画更新函数 (核心：按时间读取旋转矩阵)
# ==========================================
def update(frame):

    
    print(f"正在渲染第 {frame} 帧...") # 加这一行用来诊断
    
    row = df.iloc[frame]
    current_time = row['Time_s']
    # ... 后续代码保持不变 ...
    # 提取当前时刻的行数据
    row = df.iloc[frame]
    current_time = row['Time_s']
    
    # 提取你表头里的旋转矩阵 
    # (根据你的截图：R11, R12, R13, R21, R22, R23, R31, R32, R33)
    R = np.array([
        [row['R11'], row['R12'], row['R13']],
        [row['R21'], row['R22'], row['R23']],
        [row['R31'], row['R32'], row['R33']]
    ])
    
    # 将基础坐标系的轴向量，乘以当前时刻的旋转矩阵
    # 得到设备当前时刻在世界坐标系下的朝向
    rot_x = R @ base_x
    rot_y = R @ base_y
    rot_z = R @ base_z
    
    # 更新绘图数据：从原点 (0,0,0) 画向旋转后的终点
    x_axis.set_data_3d([0, rot_x[0]], [0, rot_x[1]], [0, rot_x[2]])
    y_axis.set_data_3d([0, rot_y[0]], [0, rot_y[1]], [0, rot_y[2]])
    z_axis.set_data_3d([0, rot_z[0]], [0, rot_z[1]], [0, rot_z[2]])
    
    # 更新时间显示文本
    time_text.set_text(f'Time: {current_time:.3f} s  (Frame: {frame}/{num_frames})')
    
    return x_axis, y_axis, z_axis, time_text

# ==========================================
# 4. 运行并播放动画
# ==========================================
print(f"开始加载动画... 共 {num_frames} 帧，每帧间隔 {interval_ms} ms")

ani = FuncAnimation(
    fig, 
    update, 
    frames=num_frames, 
    interval=interval_ms, # 使用根据 Time_s 计算出的真实时间间隔
    blit=False,
    repeat=False # 播完一遍是否重复
)

plt.show()
```
