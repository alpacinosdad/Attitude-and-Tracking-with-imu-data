```python
import pandas as pd
import matplotlib.pyplot as plt

# CSV 文件名
filename = 'BAT_Heat_Log_Data_2026_06_08_13_39_53_imu_result.csv'

# 读取 CSV
df = pd.read_csv(filename, encoding='utf-8')
df.columns = df.columns.str.strip()  # 去掉列名两端空格

# 检查列名
required = ['aax', 'aay', 'aaz']
for col in required:
    if col not in df.columns:
        raise ValueError(f'CSV 中找不到列: {col}')

# 计算 aaz-9800
df['aaz_adj'] = df['aaz'] - 9800



# 绘图
plt.figure(figsize=(10, 6))
plt.plot(df.index, df['aax'], label='aax')
plt.plot(df.index, df['aay'], label='aay')
plt.plot(df.index, df['aaz_adj'], label='aaz-9800')
plt.xlabel('样本点')
plt.ylabel('加速度')
plt.title('aax / aay / aaz-9800 可视化')
plt.legend()
plt.grid(True)
plt.tight_layout()
plt.show()
```
